```python
import cv2
import os
import time

# Define the directory where captured images will be stored
data_dir = "sign_language_dataset"
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# Define the gestures you want to capture
gestures = ['hello', 'love', 'drink']  # Add more gestures as needed

# Initialize camera
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Unable to open camera.")
    exit()

# Function to capture and save images for each gesture
def capture_images(gesture):
    gesture_dir = os.path.join(data_dir, gesture)
    if not os.path.exists(gesture_dir):
        os.makedirs(gesture_dir)

    print(f"Capturing images for {gesture}. Press 'q' to stop.")

    count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Failed to capture frame.")
            break

        cv2.imshow('Capture', frame)

        key = cv2.waitKey(1)
        if key == ord('q'):
            break
        elif key == ord('s'):
            filename = os.path.join(gesture_dir, f"{gesture}_{count}.jpg")
            cv2.imwrite(filename, frame)
            print(f"Saved: {filename}")
            count += 1

# Capture images for each gesture
for gesture in gestures:
    capture_images(gesture)

# Release resources
cap.release()
cv2.destroyAllWindows()
```
